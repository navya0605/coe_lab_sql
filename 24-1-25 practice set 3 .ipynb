```python
#task: createa line plot showing the growth of a savings bank over 6 months with following
#x-axis:months{1-6}
#y-axis:savings amount
#customizations:green dashed line with markers at each point,labels and a title
x=['Jan','Feb','Mar','Apr','May','June']
y=[1000,600,1200,1500,1350,1700]
plt.bar(x,y,color='pink')
plt.plot(x,y,linewidth=2,linestyle='dotted',color='green',marker='o')
plt.xlabel("first Half months")
plt.ylabel("savings growth")
plt.title("growth of savings account in XYZ Bank")
plt.show()
```

```python
#pie chart

lan = ['python', 'ruby', 'c++', 'java']
per = [40, 10, 20, 30]
plt.pie(per, labels=lan, colors=['violet', 'hotpink', 'lightgreen', 'aqua'],autopct='%1.1f%%',startangle=180)
plt.title("Programming Languages Percentage")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("stores.csv" , encoding="latin1")


# Preprocess to calculate monthly sales
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.strftime('%b')
monthly_sales = df.groupby('Month')['Sales'].sum()
monthly_sales = monthly_sales.reindex(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                                       'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Line plot for monthly sales
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', color='b', label='Monthly Sales')
plt.title('Monthly Sales Analysis')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.grid()
plt.legend()
plt.show()

# Histogram for order quantity distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Quantity'], bins=10, color='orange', edgecolor='black')
plt.title('Order Quantity Distribution')
plt.xlabel('Order Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Task 2: Customer Demographics
# Generate sample dataset
ages = np.random.randint(18, 65, size=200)
spendings = np.random.uniform(100, 2000, size=200)

# Histogram for age distribution
plt.figure(figsize=(10, 6))
plt.hist(ages, bins=10, color='green', edgecolor='black')
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Scatter plot for spending patterns
plt.figure(figsize=(10, 6))
plt.scatter(ages, spendings, alpha=0.7, color='purple')
plt.title('Customer Spending Patterns')
plt.xlabel('Age')
plt.ylabel('Spending Amount')
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Task 3: School Performance
# Generate sample dataset
subjects = ['Math', 'Science', 'English', 'History', 'Art']
marks = {
    'Student1': np.random.randint(50, 100, size=5),
    'Student2': np.random.randint(40, 90, size=5),
    'Student3': np.random.randint(60, 95, size=5),
    'Student4': np.random.randint(30, 85, size=5),
    'Student5': np.random.randint(45, 80, size=5),
}

marks_df = pd.DataFrame(marks, index=subjects)

# Bar chart for student marks comparison
marks_df.plot(kind='bar', figsize=(10, 6))
plt.title('Student Marks Across Subjects')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.legend(title='Students')
plt.show()

# Box plot using Matplotlib
plt.figure(figsize=(10, 6))
plt.boxplot(marks_df.values, labels=marks_df.columns, patch_artist=True, 
            boxprops=dict(facecolor='lightblue', color='blue'),
            medianprops=dict(color='red', linewidth=2))
plt.title('Distribution of Marks Across Subjects')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
